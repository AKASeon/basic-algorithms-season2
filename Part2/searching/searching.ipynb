## Hashing - 해시 탐색!
- 이전까지는 아이템이 어디에 위치하는지를 이용하여 문제해결 (순차, 이진 탐색)
- O(1)로 찾을 수 있는 자료구조 구축 - 해시 테이블
- 해쉬테이블은 컬렉션의 일종으로 추후 데이터를 찾기 쉽게 하도록 만들어짐
- 해쉬테이블의 각 자리는 슬롯(slot)이라고 칭함
- 각 슬롯 당 아이템을 가질 수 있고, 0부터 시작하는 정수로 이름 붙여짐
- 처음 해시 테이블이 만들어지면 모든 슬롯은 빈 상태 (Python에서는 None)

```python
from IPython.display import Image
Image('img/hash.png')
```

###hash function : 슬롯과 아이템을 매핑 후 슬롯 이름에 해당하는 정수를 반환.
### remainder method : 각 아이템을 해시 테이블 크기로 나눈뒤 나머지에 해당하는 값을 hash value에 지정

- [54, 26, 93, 17, 77, 31] % 11

#### Load Factor

- 전체 슬롯에서 사용중인 슬롯 비율. Load factor가 100%로 증가할수록 데이터를 찾거나 삽입하기 위해 필요한 탐사 횟수는 비약적으로(dramatically) 증가한다.
- 위 테이블의 경우 6/11

#### 복잡도
- O(1), key - value로 이뤄져 있기에 바로바로 찾아갈 수 있다

## 충돌 - Colison

- (위 그림5 참조) 77%11 = 0 이기에 77은 0번 슬롯에 지정.
- 만약 44가 아이템 리스트에 있었다면?? (44%11 = 0)
- 두 개 이상의 아이템이 같은 슬롯을 필요로 하는 상황을 Collison(or 'clash')이라고 부름

## Hash Functions
- perfect hash function : 하나의 슬롯에 하나의 아이템만 배정받는 경우
- 만약 아이템 리스트가 확정된 경우라면? -> 가능!
- 만약 아이템 리스트가 가변적이라면 ? -> 불가능!

### 해시 테이블의 크기를 늘린다면?
아이템의 수에 알맞게 테이블의 크기를 늘리면 각각의 아이템이 자기만의 슬롯에 들어가도록 할 수 있다. 이 때 발생하는 문제점은?
- 메모리 비용의 증가! ex) 25명의 학생을 위한 9자리의 사회보장번호(주민번호)를 담아둘 해시 테이블 생성시 10억개의 슬롯이 필요.
- 그렇다면 충돌의 수를 가장 최소화할 수 있는 hash function을 만드는 것이 좋지 않을까?(추가로 계산하기 쉽고, 테이블 내에 고르게 배분하는 그런 함수)

### folding method - 숫자의 각 자릿수를 더해 해시 값을 만드는 것.
- 02-6674-0565 / (2,66,74,5,65)
- 합 212, 해시 테이블의 슬롯을 11로 가정했을때 212%11 = 3
- 3번 슬롯 == 02-6674-0565

### mid-square method 
- 아이템을 제곱한 뒤 나온 숫자의 일부분을 가지고 해시 값을 만드는 방법
    - ex) 첫 아이템 44*44 = 1,936
    - 1936의 가운데 부분 93을 가지고 11로 나눠 93%11 = 5
    - 44를 5번 슬롯에 지정

## Collision Resolution - 충돌 해결하기

### Open Addressing
- 하나의 슬롯에 2개 이상의 아이템이 지정될 때, 두 번째 지정된 아이템을 어디로든 옮겨 지정해줘야함.
- 충돌을 일으킨 슬롯에서부터 처음으로 발견되는 빈 슬롯에 저장하기
- linear probing이라고도 함

list = [54,26,93,17,77,31,44,55,20] 
##### 77, 44, 55 % 11 = 0
##### 20 % 11 = 9

```python
from IPython.display import Image
Image('img/li.png')
## figure 08
```

####Open addressing, linear probing 을 사용하고 나면 그 다음에 발생할 문제에 대비 해야함
- 만약 93을 찾는다고 가정시 93%11 = 5 -> 슬롯 5번에 93이 있어 True가 반환된다.
- 그러나 20을 찾는다면 20%11 = 9 -> 슬롯 9에는 31이 있어 False가 반환되는 사태에 대비
- 9번 슬롯 다음 10번 슬롯부터 순차 탐색이 들어가야함
- 빈 슬롯이 나오기 전까지 돌려야함

#### disadvantage to linear probing
- Cluster(클러스터) : 일부 지역의 주소들을 집중적으로 반환 하는 결과로 데이터들이 한 곳에 모이는 문제
- linear probing 사용시 클러스터가 빈번히 발생.

### Quadratic Probing - Quadratic probing operates by taking the original hash index and adding successive values of an arbitrary quadratic polynomial until an open slot is found.

#### rehasing - 충돌 후 새로운 자리를 찾는
- rehash function() : newHashValue=rehash(oldHashValue)
- where rehash(pos)=(pos+1)%sizeoftable. The “plus 3” rehash can be defined as rehash(pos)=(pos+3)%sizeoftable. 
- In general, rehash(pos)=(pos+skip)%sizeoftable. 
- skip의 값이 중요. 테이블 사이즈 값은 소수가 좋다.
- 선형 탐사의 변형을 위해 쿼드라틱 프로빙이 사용됨.
- 일정한 skip 값을 사용하지 않기 위해 h+1, h+4, h+16, h+25 ...

## Open addressing comparrison
1. Linear probing : 순차적으로 탐색하며 비어있는 버켓을 찾을 때까지 계속 진행된다. 최악의 경우, 탐색을 시작한 위치까지 돌아오게 되어 종료(모든 버켓 순회)하게 된다. 캐쉬의 효율은 높으나 데이터의 클러스터링에 가장 취약한 단점이 있다.

2. Quadratic probing : 2차 함수를 이용해 탐색할 위치는 찾는다. 캐쉬의 효율과 클러스터링에 강한 능력에서 linear probing과 double hashing probing의 중간 정도 능력을 가진다. 클러스터링에 거의 영향을 받지 않는다. 또한 많은 연산량을 요구한다.

### Chaning Method - 충돌이 발생하면 각 데이터를 해당 주소에 있는 링크드 리스트에 삽입하여 문제를 해결하는 방법. 

- 충돌 문제를 다루는 두 번째 방법 
- 각각의 슬롯이 컬렉션에 대한 참조를 갖도록
- 충돌이 일어나도 각 아이템은 테이블의 그 자리에 저장됨
- 그러나 같은 장소에 아이템이 증가함에 따라 해당 슬롯에 위치한 컬렉션에서 해당 값 찾기가 어려워짐.
