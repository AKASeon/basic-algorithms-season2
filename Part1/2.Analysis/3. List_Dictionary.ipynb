#List

파이썬의 설계자들은 List 구조를 설계할 때 많은 고려와, 선택을하여야 했다. 이들은 많은 사람들이 List 구조를 사용하는 경우를 검토하여 이를 구현하였다고 한다. 일반적인 동작들에 대해 시간 복잡도를 보면 다음과 같다.


# List 아이템 인덱싱, 할당
1. 그 작업의 크기에 상관없이 O(1)


# List 아이템 추가 : append, concatenation
1. Append 메서드 : O(1) 
2. concatenation(연결 연산자, (+, extend)) : O(k) (k는 연결될 리스트가 가진 아이템의 수)

아래의 예제를 통해 다음과 같은 리스트 생성 방식을 알아 보자
1. for와 연결 연산
2. for와 append
3. 리스트 내포
4. range 함수

```python
# concatenation 연결 연산
def test1():
    l = []
    for i in range(1000):
        l = l + [i]

# append
def test2():
    l = []
    for i in range(1000):
        l.append(i)

# list comprehension, 리스트 내포
def test3():
    l = [i for i in range(1000)]

# range 내장 함수
def test4():
    l = list(range(1000))
```
