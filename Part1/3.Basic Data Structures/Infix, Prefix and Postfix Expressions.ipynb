#중위, 전위, 후위 표기법 (Infix, Prefix and Postfix Expressions)


우리가 보통 쓰는 수학 수식 표현법은 중위 표기법(Infix Expression)입니다.

A `*` B 와 같이 말이죠.

A + B * C 와 같은 수식의 경우에도 우리는 쉽게 문제를 풀 수 있습니다.

그 이유는 우리가 수학을 배우면서 일반적으로 정해진 **연산자의 우선 순위** 를 알고 있기 때문입니다.

A + B * C 와 같은 수식의 경우

우리는 대체로 A + ( B `*` C ) 와 같이 이해를 하고 문제를 풉니다.

왜냐하면 곱셈(`*`)과 나눗셈(/)은 덧셈(+)과 뺄셈(-)보다 __연산자의 우선 순위__가 높기 때문입니다.

또한 __동일한 연산자의 우선 순위__를 갖고 있는 경우 보통 왼쪽에서 오른쪽으로 순서대로 계산합니다.

만약 특정한 방법으로 연산자 우선 순위를 보장하고 싶을 경우, 우리는 __괄호__를 이용하여 원하는 순서로 수식을 계산할 수 있도록 보장합니다.

A + B `*` C + D 는 

(( A + ( B `*` C )) + D )로 표현 할 수 있습니다.


연산자의 위치에 따라 수식 표현이 달라지는데, 이를 우리는 전위, 후위 표기법(Prefix, Postfix Expressions)이라고 부릅니다.

###전위 표기법

전위 표기법은 2개의 피연사자의 앞에 연산자가 오도록 합니다.

A + B  ===>  + A B

수식을 읽는 방법은 연산자를 기준으로 오른쪽 편에 피연산자 2개가 있을 경우 먼저 계산하도록 합니다.

###후위 표기법

후위 표기법은 2개의 피연사자의 뒤에 연산자가 오도록 합니다.

A + B  ===> A B +

수식을 읽는 방법은 연산자를 기준으로 왼쪽 편에 피연산자 2개가 있을 경우 먼저 계산하도록 합니다.

### 표기법 예제

중위 표기법 |   전위 표기법  |  후위 표기법
----------- | ---------------| ---------------
  A + B     |      + A B     |    A B +
  A + B `*` C |      + A `*` B C  |   A B C `*` +
  ( A + B ) `*` C  |  `*` + A B C  |    A B + C `*`
  A + B `*` C + D | + + A `*` B C D | A B C `*` + D +
  ( A + B ) `*` ( C + D ) | `*` + A B + C D | A B + C D + `*`
  A `*` B + C `*` D | + `*` A B `*` C D | A B `*` C D `*` +
  A + B + C + D | + + + A B C D | A B + C + D +



# 중위 표기법에서 전위, 후위 표기법으로 변환

중위표기법에서 전위 혹은 후위 표기법으로 변환하기 위해서는 다음과 같은 순서로 변환을 하도록 합시다.

### 표기법 변환 방법
1. 중위 표기법에서 __연산자 우선순위__에 따라 수식에 괄호를 만든다.
2. 괄호 안의 수식을 기준으로 전위 혹은 후위 표기법으로 바꾼다.
3. 2번 과정을 모든 괄호안의 수식이 만족하도록 반복한다.
4. 더이상 전위 혹은 후위 표기법으로 바꾸지 않아도 된다면 모든 괄호를 제거한다.

#일반적인 중위 표기법에서 후위 표기법으로 변환 (알고리즘)

 A + B `*` C 를 후위 표기법으로 변환하면  A B C `*` + 입니다.
 
 위의 경우 왼쪽에서 오른쪽으로 읽었을 때, 연산자들의 순서가 역순으로 나열 된 것을 알 수 있습니다.
 
 또한 ( A + B ) `*` C 의 경우, 후위 표기법으로 변환하면   A B + C `*` 입니다.
 
 이 경우는 왼쪽에서 오른족으로 읽었을 때, 연산자 우선순위가 높은 괄호 안의 + 가 먼저 나열 된 것을 알 수 있습니다.

우리는 중위 표기법을 후위 표기법으로 변환하기 위해서, 연산자를 스택으로 보관할 것입니다.

연산자를 보관을 하되, 다음 연산자를 비교하여 

최근에 보관한 연산자의 우선순위가 높을 경우에는, 보관한 연산자를 나열합니다.

그렇지 않을 경우에는 다음 연산자를 스택에 보관합니다.

또한 왼쪽 괄호 연산자를 만났을 경우, 오른쪽 연산자를 만나기 전까지 위의 룰을 따르지만 오른쪽 괄호 연산자를 만났을 경우, 스택에 있는 연산자들을 모두 나열합니다.


이를 정리하면 다음과 같다.

1. 연산자를 보관하기 위한 opstack 이라는 빈 스택을 만든다. 출력을 위한 빈 리스트도 만든다.
2. string 메서드 "split"을 이용하여 중위 표기법으로 작성된 문자열을 리스트로 변환한다.
3. 왼쪽부터 오른쪽으로 리스트의 각 항목을 조사한다.
    - 항목이 피연사자라면, 출력을 위한 리스트에 붙인다.
    - 항목이 왼쪽 괄호라면, opstack에 넣는다.
    - 항목이 오른쪽 괄호라면, 왼쪽 괄호가 나올때까지 opstack의 항목을 계속 꺼낸다. 그리고 출력을 위한 리스트의 끝에 연산자를 붙인다. 
    - 항목이 연산자(+, -, `*`, /)라면, opstack에 넣는다. 단, 최근에 들어간 스택의 항목이 넣을 항목의 연산자보다 우선순위가 높거나 같다면, 최근에 들어간 스택의 항목을 제거하고 출력을 위한 리스트에 붙인다.
4. 입력된 문자열의 항목을 모두 조사했다면, opstack을 확인한다. 남아 있는 opstack의 연산자들을 제거하고, 제거한 연산자들을 출력을 위한 리스트에 붙인다.

```python
from pythonds.basic.stack import Stack

def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixList.append(token)
        elif token == '(':
            opStack.push(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        else:
            while (not opStack.isEmpty()) and \
               (prec[opStack.peek()] >= prec[token]):
                  postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    return " ".join(postfixList)

print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
```

1. operandStack 이라는 빈 스택을 만든다.
2. string 메서드 'split'을 이용하여 입력 문자열을 리스트형으로 바꾼다.
3. 왼쪽에서 오른쪽 방향으로 리스트의 항목을 조사한다.
    - 만약 항목이 피연산자라면, 문자를 숫자(integer)로 바꾸고 operandStack에 넣는다.
    - 만약 항목이 연산자( +, -, `*`, / )라면, operandStack에서 항목을 2개 꺼낸다. 처음으로 꺼낸 항목이 두번째 피연산자이고, 두번째로 꺼낸 항목이 첫번째 피연산자이다. 피연사자와 연산자들을 계산한 후, 그 결과를 operandStack에 넣는다.
4. 입력된 항목 조사가 끝났다면, 결과는 스택의 맨 위에 있다. operandStack에서 항목을 꺼낸 후 반환한다.

```python
from pythonds.basic.stack import Stack

def postfixEval(postfixExpr):
    operandStack = Stack()
    tokenList = postfixExpr.split()

    for token in tokenList:
        if token in "0123456789":
            operandStack.push(int(token))
        else:
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.push(result)
    return operandStack.pop()

def doMath(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    else:
        return op1 - op2

print(postfixEval('7 8 + 3 2 + /'))
```
