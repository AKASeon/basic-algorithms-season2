## search( )

```python
def search(self,item):
    current = self.head
    found = False
    stop = False
    while current != None and not found and not stop:
        if current.getData() == item:
            found = True
        else:
            if current.getData() > item:
                stop = True
            else:
                current = current.getNext()

    return found
```

- unordered list의 탐색과정은 모든 노드를 둘러본 후에야 끝!
- ordered list의 탐색은 그렇지 않다.
- 탐색 중인 노드의 값이 찾고있는 item보다 높아진 경우 바로 return False

# Analysis of Linked Lists¶
### 복잡도 확인 / 해당 메소드의 과정이 순회를 필요로 하는가?
- n개의 노드를 가진 링크드 리스트
- isEmpty() - O(1), head의 None여부만 확인하면 됨.
- 반면 size()는 O(n) n개의 노드를 순회해야만 알 수 있다.
- unordered list에서 add()는 O(1), head 옆에 바로 붙이면 되니까
- search, remove, add(in ordered list) 메소드들은 순회과정을 필요로 함. 이러한 메소드들의 복잡도는 O(n) 최악의 경우 모든 리스트의 노드를 돌아야 할 수도
