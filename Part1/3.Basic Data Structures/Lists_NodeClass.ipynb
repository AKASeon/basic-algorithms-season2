## The Unordered List Abstract Data Type¶
정렬되지 않은 리스트의 구조는 아이템 각각이 서로를 기준으로 상대적인 위치를 갖고 있는 아이템들의 집합이다.

### 리스트 만들기!
List() - 빈 리스트 만들기. return empty list.

add(item) - 리스트에 새로운 아이템 더하기. returns nothing. Assume the item is not already in the list.

remove(item) - 리스트로부터 해당 아이템 제거. Assume the item is present in the list.

search(item) - 리스트안에서 아이템 찾기. returns a boolean value.

isEmpty() - 리스트가 비었는지 아닌지 확인. returns a boolean value.

size() 리스트안에 몇 개의 아이템이 있는지 확인. returns an integer.

append(item) 리스트의 마지막에 아이템 추가. Assume the item is not already in the list.

index(item) 해당 아이템의 위치를 반환. returns an integer.

insert(pos,item) 해당 아이템을 pos 위치에 추가. Assume the item is not already in the list and there are enough existing 
items to have position pos.

pop() 리스트의 마지막 아이템을 제거 후 반환. returns item. Assume the list has at least one item.

pop(pos) 리스트의 pos위치 아이템을 제거 후 반환. returns item.

### Advantages[edit]
- Linked lists are a dynamic data structure, allocating the needed memory while the program is running.
- Insertion and deletion node operations are easily implemented in a linked list.
- Linear data structures such as stacks and queues are easily executed with a linked list.
- They can reduce access time and may expand in real time without memory overhead.

### Disadvantages[edit]
- They have a tendency to use more memory due to pointers requiring extra storage space.
- Nodes in a linked list must be read in order from the beginning as linked lists are inherently sequential access.
- Nodes are stored incontiguously, greatly increasing the time required to access individual elements within the list.
- Difficulties arise in linked lists when it comes to reverse traversing. For instance, singly linked lists are cumbersome to navigate backwards[1] and while doubly linked lists are somewhat easier to read, memory is wasted in allocating space for a back pointer.

#The Node Class

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext
```
