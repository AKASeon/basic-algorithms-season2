## 큐를 구현 하려면
다음의 구조와 연산으로 정의할 수 있다. 순서를 갖는 컬렉션에 "rear" 라고 부르는 끝에서 추가되고, "front" 라는 다른 끝에서 삭제 된다.  
- **Queue()**
    - 비어있는 새로운 큐를 만든다.
- **enqueue(item)**
    - 큐의 rear 로 새로운 아이템이 추가 
- **dequeue()**
    - 큐로부터 아이템을 제거
- **isEmpty()**
    - 큐가 비어있는지 여부를 확인
- **size()**
    - 큐 안에 있는 아이템의 개수
    
아래 Table1은 새로운 빈 큐를 만들고, 큐의 연산이 순서대로 진행되는 결과를 나타낸다.  
첫번째 enqueue 된 '4'가 dequeue에 의해 반환되는 첫번째 아이템이 된다.

#### Table1. Example Queue Operations
Queue Operation | Queue Conntents | Return Value
:--------------:|:---------------:|:------------:
q.isEmpty()     | []              | True
q.enqueue(4)    | [4]             |

## 파이썬으로 큐(Queue) 구현
- 큐로 사용할 리스트(list)에서 어느쪽 끝이 rear 인지 front 인지에 대한 결정이 필요 (아래 예에서는 리스트의 0 번째가 rear로 사용)
- insert 함수로 queue의 rear에 새로운 데이터를 추가
- pop 함수로 queue의 front 데이터를 빼냄

### Listing 1

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.insert(0, item)
        
    def dequeue(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
```

```python
q = Queue()
```

```python
q.enqueue(4)
```

```python
q.items
```

```python
q.enqueue('dog')
q.enqueue(True)
```

```python
q.items
```

```python
print(q.size())
```
